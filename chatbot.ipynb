```python
import pandas as pd
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from langchain.document_loaders import PyPDFLoader
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
from langchain.llms import HuggingFacePipeline
from flask import Flask, request, jsonify
import os
from google.colab import files

# 🔹 อ่านข้อมูลคำถามและคำตอบจากไฟล์ CSV
def load_csv_data(file_path):
    df = pd.read_csv(file_path)
    questions_and_answers = df.to_dict(orient="records")
    return questions_and_answers

# 🔹 Step 1: อ่านไฟล์ CSV
print("กรุณาอัปโหลดไฟล์ CSV ของคำถามและคำตอบ")
uploaded = files.upload()
csv_path = list(uploaded.keys())[0]
print(f"ไฟล์ที่อัปโหลด: {csv_path}")

data = load_csv_data(csv_path)  # โหลดข้อมูลจาก CSV

# 🔹 Step 2: ตั้งค่าโมเดล
model_name = "facebook/opt-1.3b"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device=-1,  # ใช้ CPU
    truncation=True,
    max_length=100,
    batch_size=1
)
llm = HuggingFacePipeline(pipeline=pipe)

# 🔹 Step 3: การค้นหาคำตอบจากข้อมูล CSV ตามคำถาม
def search_answer(question, data):
    for item in data:
        if question in item["Question"]:
            return item["Answer"]
    return "ไม่พบคำตอบที่ตรงกับคำถาม"

# 🔹 Step 4: ตั้งค่า Web API
app = Flask(__name__)

@app.route("/chat", methods=["POST"])
def chat():
    data = request.json
    question = data.get("question", "")
    answer = search_answer(question, data)  # ค้นหาคำตอบจาก CSV
    return jsonify({"answer": answer})

# 🔹 Step 5: โหมด Command-Line สำหรับถามหลายๆ ข้อ
def chat_cli():
    print("\nQ&A Chatbot (เกี่ยวกับโรคซึมเศร้า) - พิมพ์ 'exit' เพื่อออก\n")
    while True:
        question = input("คุณ: ")
        if question.lower() == "exit":
            print("ขอบคุณที่ใช้บริการน้องอุ่นใจนะคะ!")
            break
        answer = search_answer(question, data)  # ค้นหาคำตอบจาก CSV
        print(f"Chatbot: {answer}\n")

# 🔹 Step 6: เลือกโหมดการทำงาน
if __name__ == "__main__":
    choice = input("เลือกโหมด (1: Command-Line, 2: Web API): ")
    if choice == "1":
        chat_cli()  # ใช้ฟังก์ชันถามตอบหลายคำถามใน Command-Line
    elif choice == "2":
        app.run(host="0.0.0.0", port=5000)
    else:
        print("กรุณาเลือกโหมดที่ถูกต้อง")
```
